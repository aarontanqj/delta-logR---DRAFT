```python
def dlogR(res, res_b, dt, dt_b):
    """
    Computes \u0394 log R, a technique for identifying and calculating total
    organic carbon in organic-rich rocks.
    
    Inputs are:
    res = resistivity measured in ohm-m
    dt = measured transit time in \u03BCs/f
    res_b = resistivity corresponding to the \u0394dt value when the
    curves are baselined in non-source, clay-rich rocks
    
    0.02 is based on the ratio of -50\u03BCsec/ft per one resistivity
    cycle mentioned above
    """
    
    return math.log10(res / res_b) + 0.02 * (dt - dt_b)
```

```python
dlogR(res = 2.49, res_b = 1, dt = 74, dt_b = 1)
```

```python
def dlogR_den(res, res_b, den, den_b):
    """
    Computes \u0394 log R, a technique for identifying and calculating total
    organic carbon in organic-rich rocks.
    
    Inputs are:
    res = resistivity measured in ohm-m
    den = density values for the source rockin g/cm^3
    res_b = resistivity corresponding to the \u0394dt value when the
    curves are baselined in non-source, clay-rich rocks
    den_b = density values for the baseline intervals
    """
    
    return math.log10(res / res_b) - 2.5 * (den - den_b)
```

```python
dlogR_den(2.49, 1, 2.5, 1)
```

```python
def dlogR_neu(res, res_b, neu, neu_b):
    """
    Computes \u0394 log R, a technique for identifying and calculating total
    organic carbon in organic-rich rocks.
    
    Inputs are:
    res = resistivity measured in ohm-m
    neu = neutron porosity reading from the well log in the interval
    of interest
    res_b = resistivity corresponding to the \u0394dt value when the
    curves are baselined in non-source, clay-rich rocks
    neu_b = neutron porosity baseline value (scaled in fractional porosity)
    """
    
    return math.log10(res / res_b) + 4 * (neu - neu_b)
```

```python
dlogR_neu(2.49, 1, 0.3, 0.1)
```
